## Making changes to data 

```python
# To find the total 
pokemon['Total']= pokemon['HP'] + pokemon['Attack'] + pokemon['Defense'] + pokemon['Sp. Atk'] + pokemon['Sp. Def'] + pokemon['Speed']
pokemon.head()
```

```python
#dropping a column
pokemon.drop(columns=['Total'], inplace=True)
pokemon
```

```python
#how to add a column
pokemon['Total']=pokemon.iloc[:, 4:10].sum(axis=1)
pokemon.head()
```

```python
#how to reorder columns

columns= list(pokemon.columns.values)
pokemon=pokemon[columns[0:4] + [columns[-1]] + columns [4:12]] 
pokemon.head()
```

```python
df
```

```python
df.groupby(['Type 1']).mean().sort_values('Speed', ascending=False)
```

```python
df.groupby(['Type 1']).sum()
df.groupby(['Type 1']).count()
```

## Working with large amounts of data

```python
new_df = pd.DataFrame(columns= df.columns)

for df in pd.read_csv('Modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
    print(new_df)
```
